# Métodos para crear un dataframe

## 1. **Creando dataFrame usando arrays**

### ***Opción 1***

```python
# CREANDO UN ARRAY

# escribo listas dentro del array
# [1,4] --> primer fila, [2,5] --> segunda fila...
data = np.array([[1,4], [2,5], [3,4]])
```

```python
# CREANDO UN DATAFRAME
# creo un dataframe con el array anteriormente creado
df = pd.DataFrame(data, index=['fila1', 'fila2', 'fila3'], columns=['col1','col2'])
```

```python
#mostrando el dataframe
df
```

### ***Opción 2***

```python
# CREANDO UN ARRAY EN FORMA DE LISTA
data = [[1,4], [2,5], [3,4]] #lista de lista
```

```python
# CREANDO UN DATAFRAME
# creo un dataframe con el array anteriormente creado
df = pd.DataFrame(data, index=['fila1', 'fila2', 'fila3'], columns=['col1','col2'])
```

```python
# MOSTRANDO EL DATAFRAME
df
```

## 2. **Creando dataFrame usando diccionarios**

```python
# Listas usadas para el ejemplo
estados = ['California', 'Texas', 'Florida', 'Nueva York']
poblacion = [39613493, 29730311, 21944577, 19299981]
```

```python
# Almacenando listas en un diccionario
dic = {'Estados' : estados, 'Poblacion' : poblacion}
```

```python
# Creando el dataFrame
df_poblacion = pd.DataFrame(dic)
```

```python
# Mostrando el dataFrame
df_poblacion
```
